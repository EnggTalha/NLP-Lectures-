# Keras Basics

This notebook introduces the basics of Keras, a powerful and easy-to-use deep learning framework. We'll cover how to build, compile, train, and evaluate neural networks using Keras.

---

## 1. Introduction to Keras

Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow. It supports rapid experimentation and is user-friendly, modular, and extensible.

---

## 2. Importing Necessary Libraries

Let's start by importing the necessary libraries.


```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

---

## 3. Creating a Simple Neural Network

We will build a simple feedforward neural network using Keras.

### Steps:
1. Define the model architecture.
2. Compile the model.
3. Train the model.
4. Evaluate the model.

---

### 3.1 Define the Model Architecture

We'll use the `Sequential` API for simplicity.


```python
# Code Cell
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

---

### 3.2 Compile the Model

Before training, we need to configure the model with:
- Optimizer: Determines how the model weights are updated during training.
- Loss Function: Measures how well the model predicts.
- Metrics: Used to evaluate the performance.

Let's compile the model.


```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

---

### 3.3 Training the Model

We need some dummy data for training. We'll generate random data and split it into training and validation datasets.

- **Epochs:** Number of times the model sees the entire dataset.
- **Batch Size:** Number of samples processed before updating the model.



```python
# Generate dummy data
X_train = np.random.rand(1000, 10)
y_train = np.random.randint(0, 2, size=(1000, 1))

X_val = np.random.rand(200, 10)
y_val = np.random.randint(0, 2, size=(200, 1))

# Train the model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=32)
```

---

### 3.4 Evaluating the Model

After training, we evaluate the model on the validation dataset to check its performance.


```python
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")
```

---

## 4. Visualizing Training Progress

We can visualize the loss and accuracy over epochs using `matplotlib`.


```python
# Code Cell
import matplotlib.pyplot as plt

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
